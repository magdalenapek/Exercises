## Part 1

Create the bar plot seen in the image below.  It visualizes the top 10 artists who have spent the most weeks at number 1 in the dataset.

* Notice the customized title
* The figure is 8x6
* Add in the x-axis label
* The plot uses the ggplot matplotlib theme
* The bars have a black edge with a width of 3.  You'll need to add that in yourself!



```python
import pandas as pd
import matplotlib.pyplot as plt
billboard = pd.read_csv("C:/Users/magda/OneDrive/Pulpit/DataAnalysisCourseMaterials/DataAnalysis/data/billboard_charts.csv", parse_dates = (["date"]))
billboard
```

```python
plt.style.use("ggplot")
```

```python
#It visualizes the top 10 artists who have spent the most weeks at number 1 in the dataset.
```

```python
artist_rank = billboard[billboard["rank"] == 1]
```

```python
artist_rank["artist"].value_counts().head(10).plot(kind = "barh", title = "Artist With The Most Weeks At #1", figsize = (8,6), edgecolor ="black",linewidth = 3)
plt.xlabel("Weeks at 1#")
```

## Part 2

Please create the pie chart seen in the image below.  It visualizes the various artists who charted with the song "Please Mr. Postman".  

* Add in the custom title
* Use this list of colors: `colors = ["#6c5ce7","#00cec9", "#74b9ff"]`

## Part 3
Create the line plot seen in the image below. It visualizes the performance of the song "Blinding Lights"

* The figure is 9x5
* The line color is `#a29bfe`
* To invert the y-axis (so that rank 1 is the highest value), use: `plt.gca().invert_yaxis()`



```python
lights = billboard[billboard["song"] == "Blinding Lights"]
lights.set_index("date")["rank"].plot(kind = "line", 
                                      figsize = (9,5), 
                                      color = "#a29bfe",
                                     title = "Blinding Lights Chart Performance")
plt.gca().invert_yaxis()
```

## Part 4

Create the line plot seen in the image below.  It shows the chart performance (rank) of the following songs:

* All I Want For Christmas Is You by Mariah Carey
* Rockin' Around The Christmas Tree by Brenda Lee
* Jingle Bell Rock by Bobby Helms

**The date range spans from 2016-12-25 to 2021-01-01**

Notice the customized x-axis tick marks, the legend, the title, and the axis labels! Also the figure is `10x7`

To invert the y-axis, use `plt.gca().invert_yaxis()`



```python
mariah = billboard[(billboard["song"] == "All I Want For Christmas Is You") & 
                   (billboard["artist"] == "Mariah Carey") &
                   (billboard["date"].between('2016-12-25','2021-01-01'))].set_index("date")
mariah["rank"].plot()
```

```python
brenda = billboard[(billboard["song"] == "Rockin' Around The Christmas Tree") 
                   & (billboard["artist"] == "Brenda Lee") &
                   (billboard["date"].between('2016-12-25','2021-01-01'))].set_index("date")
brenda
```

```python
bobby = billboard[(billboard["song"] == "Jingle Bell Rock") 
                  & (billboard["artist"] == "Bobby Helms") &
                  (billboard["date"].between('2016-12-25','2021-01-01'))].set_index("date")
bobby
```

```python
plt.figure(figsize = (10,7))
mariah["rank"].plot(label = "All I Want For Christmas")
bobby["rank"].plot(label = "Jingle Bell Rock")
brenda["rank"].plot(label = "Rockin' Around The Christmas Tree", ylabel = "Chart Rank")
plt.gca().invert_yaxis()
plt.ylabel("Chart Rank")
plt.title("Christmas songs")
plt.legend()
plt.xticks(["2016-12-25", "2017-12-25","2018-12-25","2019-12-25","2020-12-25"],
           labels = ["Xmas 2016", "Xmas 2017", "Xmas 2018", "Xmas 2019", "Xmas 2020"])
```

## Part 5

Create the visualization seen in the image below.  It shows the top 8 songs that spent the most time on the Billboard charts (at any position) for each of the 4 Beatle's solo acts:
* John Lennon
* Wings
* George Harrison
* Ringo Starr

The details:
* 2 rows, 2 columns
* Figure is 14x8
* The main title has a fontsize of 20
* Notice the colors of each plot and its corresponding title!
* Colors are: 
    * red: #fc5c65
    * orange: #fd9644
    * blue: #2bcbba
    * green: #26de81
    


```python

    Colors are:
        red: #fc5c65
        orange: #fd9644
        blue: #2bcbba
        green: #26de81
```

```python
lennon = billboard[billboard["artist"] == "John Lennon"]["song"].value_counts()
lennon
```

```python
wings = billboard[billboard["artist"] == "Wings"]["song"].value_counts().head(8)
wings
```

```python
geo = billboard[billboard["artist"] == "George Harrison"]["song"].value_counts().head(8)
geo
```

```python
ringo = billboard[billboard["artist"] == "Ringo Starr"]["song"].value_counts().head(8)
ringo
```

```python
fig, axs = plt.subplots(2,2, figsize = (14,8))
plt.suptitle("Beatles Solo Songs (By Week On Chart)", fontsize = 20)
lennon.plot(kind = "barh", ax = axs[0][0], color = "#fc5c65")
axs[0][0].set_title("John Lennon", color = "#fc5c65")
wings.plot(kind = "barh", ax= axs[0][1], color = "#fd9644")
axs[0][1].set_title("Wings (Paul)", color = "#fd9644")
geo.plot(kind = "barh", ax = axs[1][0], color = "#2bcbba")
axs[1][0].set_title("George Harrison", color = "#2bcbba")
ringo.plot(kind = "barh", ax = axs[1][1], color = "#26de81")
axs[1][1].set_title("Ringo Starr", color = "#2bcbba")
plt.tight_layout()
```
